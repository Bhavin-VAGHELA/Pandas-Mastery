### 1. List Into Series

```python
import pandas as pd
```

```python
list = ['bhavin','kishan','hardik','gautam']
```

```python
series = pd.Series(list)
print(series)
print (type(series))
```

### 2. Integer Into Series

```python
runs = [12,13,45,16,15,18,19,20,12]
```

```python
runs_series = pd.Series(runs)
print(runs_series)
print (type(runs_series))
```

### 3. Custom Index

```python
marks = [1,2,3,4,5]
subjects = ['Hindi','Science','Geography','Maths','Chemistry']
custom_index = pd.Series(marks,index=subjects)
print(custom_index)
print(type(custom_index))
```

### 4. Setting Name Of Object In Series

```python
marks = [1,2,3,4,5]
subjects = ['Hindi','Science','Geography','Maths','Chemistry']
custom_index = pd.Series(marks,index=subjects,name='Marks Of Bhavin')
print(custom_index)
print(type(custom_index))
```

### 5. Setting Name Of Object In Series

```python
marks = {
    'sci':100,
    'maths': 200,
    'geography': 300
}

dict_series = pd.Series(marks,name='Dictionary of marks')
print(dict_series)
print(type(dict_series))
```

### 8. Series Methods

```python
# head and tail
df = pd.read_csv('bollywood.csv',index_col='lead')
movies_series = df['movie']
print(movies_series)
print(type(movies_series))
```

```python
# head
# movies_series.head() -> first 5 rows by default
# movies_series.head(10) -> first 10 rows of dataset
```

```python
# tail
# movies_series.tail() -> first 5 rows by default
# movies_series.tail(10) -> first 10 rows of dataset
```

```python
# sample -> any random row/rows from dataset
# movies_series.sample() -> any one random movie name
# movies_series.sample(5) -> any random 5 movies from dataset
```

```python
# value_counts 
movies_series.value_counts()
```

```python
# sort
movies_series.sort_values()
# to make make in descending use "ascending=False"
# for doing permanent changes use "inplace=True"
```

```python
movies_series.sort_index()
# to make make in descending use "ascending=False"
# for doing permanent changes use "inplace=True"
```

### 9. Series Math Method

```python
series = pd.read_csv('subs.csv')
subs_series = series.squeeze()
print(subs_series)
print(type(subs_series))
```

```python
# min -> gives Minimum Value From Dataset
subs_series.min()
```

```python
# max -> gives Maximum Value From Dataset
subs_series.max()
```

```python
# mean , median , mode , std , var -> statistical methods
print('The Mean Of The Cloumn Is:-')
print(subs_series.mean())
print()
print('The Median Of The Cloumn Is:-')
print(subs_series.median())
print()
print('The Mode Of The Cloumn Is:-')
print(subs_series.mode())
print()
print('The Standard deviation Of The Cloumn Is:-')
print(subs_series.std())
print()
print('The Variance Of The Cloumn Is:-')
print(subs_series.var())
```

```python
# count
subs_series.count()
```

```python
# sum 
subs_series.sum()
```

```python
# describe -> short statistical info about dataset
subs_series.describe()
```
